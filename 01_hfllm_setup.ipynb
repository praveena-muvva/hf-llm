```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've always ignored pushing my work to git",
        "I'm correcting it now",
    ]
)
```

```python
from google.colab import userdata
userdata.get('huggingface')

from transformers import AutoTokenizer

# this is the default checkpoint for sentiment-analysis pipeline
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
```

```python
raw_inputs = [
    "I've always ignored pushing my work to git",
    "I'm correcting it now",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
print(inputs)
```

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)
print(outputs.logits) #Raw unnormaliws scores
```

```python
import torch

#Coverting the logits to probabilities
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)

# {0: 'NEGATIVE', 1: 'POSITIVE'}
```
